```python
#Function finds all the candidate pairs
def findCandidatePairs(corpus_path,measure='jaccard'):
    shingle_doc_matrix, signature_matrix=getSignatureMatrix(corpus_path)
    candidate_matrix=[]
   
    no_of_docs=len(signature_matrix[0])
   
    for i in range(0,no_of_docs):
        lis=[]
        for j in range(0,no_of_docs):
            lis.append(0)
        candidate_matrix.append(lis)
   
    r=10
   
    #Jaccard Coefficient
    if(measure=='jaccard'):
        t=0.65
        for i in range(0,no_of_docs):
            for j in range(i+1,no_of_docs):
                sig_i=[row[i] for row in signature_matrix]
                sig_j=[row[j] for row in signature_matrix]
                for b in range(0,len(sig_i),r):
                    band_i = set(sig_i[b:b+r])
                    band_j = set(sig_j[b:b+r])
                    jaccard=len(band_i.intersection(band_j))/len(band_i.union(band_j))
                    #print(i,j,jaccard)
                    if jaccard>t:
                        candidate_matrix[i][j]=1
                        candidate_matrix[j][i]=1
   
       
    #Hamming Distance
    elif(measure=='hamming'):
        for i in range(0,no_of_docs):
            for j in range(i+1,no_of_docs):
                sig_i=[row[i] for row in signature_matrix]
                sig_j=[row[j] for row in signature_matrix]
                for b in range(0,len(sig_i),r):
                    band_i = sig_i[b:b+r]
                    band_j = sig_j[b:b+r]
                    hamming=0
                    for x in range(0,len(band_i)):
                        if(band_i[x]!=band_j[x]):
                            hamming+=1
                    #print(i,j,hamming)
                    if(hamming<=7):
                        candidate_matrix[i][j]=1
                        candidate_matrix[j][i]=1
       
    #Cosine Distance
    elif(measure=='cosine'):
        for i in range(0,no_of_docs):
            for j in range(i+1,no_of_docs):
                sig_i=[row[i] for row in signature_matrix]
                sig_j=[row[j] for row in signature_matrix]
                for b in range(0,len(sig_i),r):
                    band_i = sig_i[b:b+r]
                    band_j = sig_j[b:b+r]
                    cos=0
                    for x in range(0,len(band_i)):
                        if(band_i[x]==band_j[x]):
                            cos+=1
                    cos=cos/10
                    print(i,j,cos)
                    if(cos>=0.5):
                        candidate_matrix[i][j]=1
                        candidate_matrix[j][i]=1
                   
    return shingle_doc_matrix, candidate_matrix  
```
